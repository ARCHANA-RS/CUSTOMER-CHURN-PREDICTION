## correlation matrix

```python
dataset.apply(lambda x: pd.factorize(x)[0] if x.dtype!='float64' else x)
# since the monthly and total charges are in numerical so there is no need to change their datatype.
#apply() applies a function to each column , lambda x anonymous func works on each column , if not numeric then encode it, factorize - text to integers
```

```python
plt.figure(figsize=(13,8))
# creates a  new fig , size of the plot to 13 inches wide  8 inches high
corr= dataset.apply(lambda x: pd.factorize(x)[0] if x.dtype!='float64' else x).corr()
# Applies a function,  Converts categorical columns into numeric (yes/no)-0,1 ,Calculates the correlation matrix
mask= np.triu(np.ones_like(corr,dtype=bool))
# Creates a matrix,only the upper triangle,hide the upper triangle  in the heatmap .
sns.heatmap(corr,mask=mask,cmap='coolwarm',vmax=1,vmin=-1,center=0,annot=True,fmt='.2f')
# The correlation matrix you created,Sets the color gradient from blue (negative) to red (positive), ranges from -1 to 1,Centers the color map at 0,Adds numbers inside the cells.
#fmt='.2f': Formats the numbers to 2 decimal places.
```

```python
x=dataset.drop(columns=['Churn'])
y=dataset['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8, random_state=42, stratify =y)
# as there is an imbalancement in the data (churn) this stratify heps us to split the data in an equal manner.
```

## MODEL TRAINING

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# feature scaling
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled,y_train)
y_predict=knn.predict(x_test_scaled)
print("accuracy:",accuracy_score(y_test,y_predict))
print("\n confusion matrix:\n",confusion_matrix(y_test,y_predict))
print("\n classification report\n",classification_report(y_test,y_predict))
```

```python
# 855: True Negatives (TN)  predicted not churn, actually not churn

# 178: False Positives (FP)  predicted churn, actually not churn

# 175: False Negatives (FN)  predicted not churn, actually churn

# 199: True Positives (TP)  predicted churn, actually churn


# Precision = TP / (TP + FP)
# For class 1 (churn):
# 0.53 means: Out of all customers predicted as churn, 53% actually churned.

# Recall = TP / (TP + FN)
#For class 1 (churn):
# 0.53 means: Out of all actual churned customers, 53% were correctly predicted.

# Harmonic mean of precision and recall.(F1 score)
# Balances false positives and false negatives.
# For class 1 (churn): 0.53 indicates moderate performance.


# Accuracy: 0.749  ~75% of total predictions are correct.
# Precision: 0.68
# Recall: 0.68
# F1-Score: 0.68
```
